```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import preprocessing
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.feature_extraction import text
import sys
def RunClassifier(name, transformer, classifier):
    print(name)
    pipe = make_pipeline(transformer, classifier)
    print("cross-validation:")
    print(np.average(cross_val_score(pipe, X_train, y_train, cv=3,scoring="f1_macro")))
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    report = classification_report(y_test, y_pred)
    print(report)

# read data
from google.colab import drive
drive.mount('/content/drive')
orig_stdout = sys.stdout
fildHandle = open('/content/drive/MyDrive/SentimentTest/out.txt', 'w')
print(fildHandle)
print("helloword")
sys.stdout = fildHandle

df = pd.read_csv('/content/drive/MyDrive/SentimentTest/Log20210916.csv')
text =  df["description"]
textScore = df["status"]
labelEncoder = preprocessing.LabelEncoder()
df["statuslabel"]=labelEncoder.fit_transform(textScore)

# make train and test data
X_train, X_test, y_train, y_test = train_test_split(text, df["statuslabel"], random_state=7524,train_size=0.80)

# Tfidf transform
max_df = 1.0 # min word frequency
min_df = 2 # max word frequency
vect = CountVectorizer(max_df=max_df,min_df = min_df,stop_words='english' )
#vect = CountVectorizer(ngram_range=(1,2),stop_words='english')
#vect = TfidfVectorizer(max_df=max_df,min_df = min_df,stop_words='english',use_idf=False)
#vect = TfidfVectorizer(max_df=max_df,min_df = min_df,stop_words='english')

term_matrix = pd.DataFrame(vect.fit_transform(X_train).toarray(), columns=vect.get_feature_names_out())

kkk=12345
# Bayes classifier
countVector = CountVectorizer(max_df=max_df,min_df = min_df)
RunClassifier("Bayes|CountVectorizer",countVector,MultinomialNB())
# RandomForest Classifier#
RunClassifier("RandomForest|CountVectorizer",countVector,RandomForestClassifier())
# LogisticRegression Classifier#
RunClassifier("LogisticRegression|CountVectorizer",countVector,LogisticRegression())
# SVM Classifier#
RunClassifier("SVC|CountVectorizer",countVector,SVC())

# Bayes classifier
countVectorEng = CountVectorizer(max_df=max_df,min_df = min_df,stop_words='english')
RunClassifier("Bayes|CountVectorizer|stop_words",countVectorEng,MultinomialNB())
# RandomForest Classifier#
RunClassifier("RandomForest|CountVectorizer|stop_words",countVectorEng,RandomForestClassifier())
# LogisticRegression Classifier#
RunClassifier("LogisticRegression|CountVectorizer|stop_words",countVectorEng,LogisticRegression())
# SVM Classifier#
RunClassifier("SVC|CountVectorizer|stop_words",countVectorEng,SVC())

# Bayes classifier
tfidf = TfidfVectorizer(max_df=max_df,min_df = min_df)
RunClassifier("Bayes|TfidfVectorizer",tfidf,MultinomialNB())
# RandomForest Classifier#
RunClassifier("RandomForest|TfidfVectorizer",tfidf,RandomForestClassifier())
# LogisticRegression Classifier#
RunClassifier("LogisticRegression|TfidfVectorizer",tfidf,LogisticRegression())
# SVM Classifier#
RunClassifier("SVC|TfidfVectorizer",countVectorEng,SVC())

# Bayes classifier
tfidfEng = TfidfVectorizer(max_df=max_df,min_df = min_df,stop_words='english')
RunClassifier("Bayes|TfidfVectorizer|stop_words",tfidfEng,MultinomialNB())
# RandomForest Classifier#
RunClassifier("RandomForest|TfidfVectorizer|stop_words",tfidfEng,RandomForestClassifier())
# LogisticRegression Classifier#
RunClassifier("LogisticRegression|TfidfVectorizer|stop_words",tfidfEng,LogisticRegression())
# SVM Classifier#
RunClassifier("SVC|TfidfVectorizer|stop_words",tfidfEng,SVC())

# Bayes classifier
tfNonIdf = TfidfVectorizer(max_df=max_df,min_df = min_df,use_idf=False)
RunClassifier("Bayes|TfidfVectorizer|tfNonIdf",tfNonIdf,MultinomialNB())
# RandomForest Classifier#
RunClassifier("RandomForest|TfidfVectorizer|tfNonIdf",tfNonIdf,RandomForestClassifier())
# LogisticRegression Classifier#
RunClassifier("LogisticRegression|TfidfVectorizer|tfNonIdf",tfNonIdf,LogisticRegression())
# SVM Classifier#
RunClassifier("SVC|TfidfVectorizer|tfNonIdf",tfNonIdf,SVC())

# Bayes classifier
tfNonIdfEng = TfidfVectorizer(max_df=max_df,min_df = min_df,stop_words='english',use_idf=False)
RunClassifier("Bayes|TfidfVectorizer|tfNonIdf|stop_words",tfNonIdfEng,MultinomialNB())
# RandomForest Classifier#
RunClassifier("RandomForest|TfidfVectorizer|tfNonIdf|stop_words",tfNonIdfEng,RandomForestClassifier())
# LogisticRegression Classifier#
RunClassifier("LogisticRegression|TfidfVectorizer|tfNonIdf|stop_words",tfNonIdfEng,LogisticRegression())
# SVM Classifier#
RunClassifier("SVC|TfidfVectorizer|tfNonIdf|stop_words",tfNonIdfEng,SVC())


sys.stdout = orig_stdout
fildHandle.close()

```
